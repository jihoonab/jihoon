```python
images='/content/drive/MyDrive/cat (1).jpg'
image=cv2.imread(images)
```

cat 이미지를 imread 함수를 이용하여 불러온다


```python
cv2_imshow(image)
```

불러온 이미지를 출력한다

```python
canny=cv2.Canny(image, 100, 255)
cv2_imshow(canny)
```

cv2.canny는 이미지의 엣지 검출 함수로 이미지의 엣지만 되돌려 준다.
canny 함수를 사용하여 이미지의 테두리만 출력한다
